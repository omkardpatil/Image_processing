```python
from google.colab.patches import cv2_imshow # for image display
import cv2
import numpy as np
import pandas as pd
```

```python
img= cv2.imread('/content/img_nature.jpg')
img_copy = img.copy()
cv2_imshow(img)
print("image shape :" + str(img.shape))
```

# Manual median filter

```python
def replace_by_median(img):
  # take a subarray, leave border
  # flatten it
  # find median
  # replace at appropriate position
  for x in range(1,img.shape[0]-2):
    for y in range(1,img.shape[1]-2):
      for z in range(3):
        temp_arr=img[x-1:x+2,y-1:y+2,z]
        # print(temp_arr)
        # return img
        temp_arr = temp_arr.flatten()
        # print(temp_arr)
        img[x,y,z]=np.median(temp_arr)
        # print(img[x,y,z])
        # print(np.median(temp_arr))
  return img
```

```python
img_median = replace_by_median(img)
cv2_imshow(img_median)
```

```python
def median_filter(src):
  dst = cv2.medianBlur(src, 3)
  return dst
```

```python
img=img_copy.copy()
cv2_imshow(img)
img_new = median_filter(img)
cv2_imshow(img_new)
```
